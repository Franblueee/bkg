## **Download MedQA LLM model**
We can download a pre-trained LLM from hugging face using the *transformers* library. The models are quite large and can take some time to download depending on your computer.

```python
from transformers import LlamaConfig

config = LlamaConfig.from_pretrained("Henrychur/MMed-Llama-3-8B-EnIns")
```

```python
import transformers

tokenizer = transformers.AutoTokenizer.from_pretrained(
    "Henrychur/MMed-Llama-3-8B-EnIns",
    model_max_length=2048,
    use_fast=False,
    trust_remote_code=True
)
```
